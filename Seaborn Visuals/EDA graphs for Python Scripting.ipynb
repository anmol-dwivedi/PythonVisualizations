# Joint Grid

```python
g = sns.JointGrid(data=df, x="Engine Capacity(cc)", y="BrakeHorsePower (bhp)")
g.plot_joint(sns.scatterplot, linewidth=1)
g.plot_marginals(sns.histplot, kde=True)
```

```python
g = sns.JointGrid(data=df, x="Engine Capacity(cc)", y="BrakeHorsePower (bhp)")
g.plot(sns.regplot, sns.boxplot)
```
